```python
import pandas as pd

## åˆ›å»ºDataFrame ##
df = pd.DataFrame({'aaa':[1,2,3], 'bbb':[4,5,6], 'ccc':[7,8,9]})
print(df)
```

### å–è¡Œ

å–åˆ—çš„æ—¶å€™å› ä¸ºæœ‰åˆ—åï¼Œä½†æ˜¯æ¯è¡Œä¸èƒ½éƒ½æ‹¥æœ‰ä¸€ä¸ªåå­—ï¼Œæ‰€ä»¥ç”¨ç´¢å¼•é¡ºåºæ¥å–è¡Œã€‚

```python
print('å–ç¬¬ä¸€è¡Œ')
line = df[:1]
print(line)
print('-------------------')

print('å–ç¬¬ä¸€è¡Œ')
line = df[0:1]
print(line)
print('-------------------')

print('å–å‰ä¸¤è¡Œ')
line = df[:2]
print(line)
print('-------------------')

print('å–æ‰€æœ‰çš„è¡Œ')
line = df[:]
print(line)

print('-------------------')
print('å–å€’æ•°ç¬¬ä¸€è¡Œ')
line = df[-1:]
print(line)
```

## è¡Œè¿æ¥ï¼ˆçºµå‘ï¼šå˜é•¿ï¼‰- concat

```python
df1=pd.DataFrame({'key':['a','b','c'],'value':[1,2,3]})  
print(df1)
print('---------------------')
df2=pd.DataFrame({'key':['d','e','f'],'value':[4,5,6]}) 
print(df2)
print('---------------------')
df3=pd.concat([df1,df2])
print(df3)
```

## åˆ—è¿æ¥ï¼ˆæ¨ªå‘ï¼šå˜å®½ï¼‰- merge

```python
df1=pd.DataFrame({'key':['a','b','c'],'value1':[1,2,3]})
print(df1)
print('---------------------')
df2=pd.DataFrame({'key':['a','b','c'],'value2':[4,5,6]}) 
print(df2)
print('---------------------')
df3=pd.merge(df1,df2)
print(df3)
```
